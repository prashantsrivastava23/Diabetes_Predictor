```python
from sklearn.tree import DecisionTreeClassifier
from pandas import read_csv
import pandas as pd
import numpy as np
import random as rnd
from sklearn.metrics import accuracy_score
```

```python
filename = 'diabetes.csv'     
names = ['preg','plas','pres','skin','test','mass','pedi','age','class']
data = read_csv(filename, sep=",") 
array = data.values
X = array[:,0:8]
Y = array[:,8]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=1234)
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, Y_train)
```

```python
rnd.seed(123458)
X_new = X[rnd.randrange(X.shape[0])]
X_new = X_new.reshape(1,8)
YHat = model.predict(X_new)
```

```python
df = pd.DataFrame(X_new, columns = names[:-1])
df["predicted"] = YHat
df.head()
```

```python
YHat = model.predict(X_test)
print(round(accuracy_score(Y_test, YHat)*100, 2))
```

```python
model = DecisionTreeClassifier(criterion='entropy', max_leaf_nodes=20)
dt = model.fit(X_train, Y_train)
YHat = model.predict(X_test)
```

```python
print(round(accuracy_score(Y_test, YHat)*100, 2))
```
